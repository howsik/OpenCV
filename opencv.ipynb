### Addition or Brightness

```python
# reading image
img_bgr = cv2.imread(r'C:\Users\acer\Downloads\images\New_Zealand_Coast.jpg',cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display 18x18 pixel image.
Image(filename=r'C:\Users\acer\Downloads\images\New_Zealand_Coast.jpg')
```

```python
# brightness
# putting image in numpy ones and multiply by 50
matrix = np.ones(img_rgb.shape,dtype = "uint8") * 50
# adding, subracting rgb_image with matrix using cv2.add and cv2.subtract
img_rgb_brighter = cv2.add(img_rgb,matrix)
img_rgb_darker = cv2.subtract(img_rgb,matrix)
# displaying the image
plt.figure(figsize=(18,5))
plt.subplot(131); plt.imshow(img_rgb_darker);  plt.title("Darker")
plt.subplot(132); plt.imshow(img_rgb); plt.title('original')
plt.subplot(133); plt.imshow(img_rgb_brighter);plt.title("Brighter")

```

### Multiplication or Contrast

```python
# putting image in numpy ones and multiply by some factor
matrix1 = np.ones(img_rgb.shape) * .8
matrix2 = np.ones(img_rgb.shape) * 1.2

# multiplying float img_rgb with matrix using cv2.multiply method and 
#finaly convert it into uint8

img_rgb_darker   = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_brighter = np.uint8(cv2.multiply(np.float64(img_rgb), matrix2))

# displaying the image
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_darker);  plt.title("Lower Contrast");
plt.subplot(132); plt.imshow(img_rgb);         plt.title("Original");
plt.subplot(133); plt.imshow(img_rgb_brighter);plt.title("Higher Contrast");
```

```python
# here we solving the issue in previous higher contrast 
# because of multiplying values with higher values it become larger than 255
# so we are using clip method
img_rgb_higher  = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), matrix2),0,255))
plt.imshow(img_rgb_higher);plt.title("Higher Contrast");
```

#### Read and write videos

```python
# read video from computer
source = r'C:\Users\acer\Downloads\test_video.mp4'
cap = cv2.VideoCapture(source)
```

```python
# checking a video is opened
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
```

```python
# read and display one frame
#ret,frame = cap.read()
#plt.imshow(frame[...,::-1])
```

```python
# Read until video is completed
while(cap.isOpened()):
      
# Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
    # Display the resulting frame
        cv2.imshow('Frame', frame)
          
    # Press Q on keyboard to exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
# break loop          
    else:
        break
        
# When everything done, release
# the video capture object
cap.release()
```

```python
cv2.destroyAllWindows()
```

```python
# writeing video
# Default resolutions of the frame are obtained.
# Convert the resolutions from float to integer.
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Define the codec and create VideoWriter object.
out_avi = cv2.VideoWriter(r'D:\opencv\out_1.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))

out_mp4 = cv2.VideoWriter(r'D:\opencv\out_2.mp4',cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width,frame_height))
```

```python
# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
    
  if ret == True:
    
    # Write the frame to the output files
    out_avi.write(frame)
    out_mp4.write(frame)
    
  # Break the loop
  else: 
    break
```

```python
# When everything done, release the VideoCapture and VideoWriter objects
cap.release()
out_avi.release()
out_mp4.release()
```
